Sets (Conjuntos) en Python

Los sets son colecciones no ordenadas, de elementos unicos, estas se definen utilizando llaves {} 
o la función set(), estos son mutables, no permiten elementos duplicados y no se pueden acceder por índice.

Si existen elementos duplicados no arrojan un error, simplemente se ignoran 
y no se agregan al set.

```python
frutas = {'manzana', 'banana', 'naranja', 'manzana'}
print(frutas)
print(type(frutas))

# La longitud del set es 3, ya que 'manzana' se repite y se ignora 
print(len(frutas)) 
```

Un set puede contener diferentes tipos de datos INMUTABLES (hashables), 
como enteros, strings, decimales, booleanos, tuplas (con elementos hashables) 
y frozensets, además, puede ser modificado después de su creación.

No pueden contener listas, diccionarios u otros sets, ya que estos NO son 
hashables (son mutables o no tienen hash definido).

```python
mixed_set = {1, "hola", 3.14, True} # True se interpreta como 1, por lo que no se agrega al set
print(mixed_set)
print(type(mixed_set))

# Recorriendo un set

for item in mixed_set:
    print(item) # La impresión no es ordenada

# Buscar elementos en un set
if 1 in mixed_set:
    print('Founded!')

# Metodos en un set

aws_services = set() # Set vacio

# Add

aws_services.add('Route 53')
print(aws_services)

# Update
# Al actualizar podemos agregar el contenido de otras listas, tuplas o conjuntos

new_services = {'EC2', 'ECC', 'S3'}
aws_services.update(new_services)
print(aws_services)

# Delete - Discard - Pop - Clear
```

Delete comprueba que el elemento a remover exista, si no existe, lanzara un error

```python
aws_services.remove('ECC')
print(aws_services)
```

Discard, si no encuentra el elemento a eliminar simplemente mostrara los elementos restantes

```python
aws_services.discard('ECC')
print(aws_services)
```

Pop elimina un elemento aleatorio del set y lo devuelve, si el set esta vacio, lanzara un error

```python
aws_services.pop()
print(aws_services)
```

Clear elimina todos los elementos del set, dejando un set vacio

```python
aws_services.clear()
print(aws_services)


# Operaciones de conjuntos

a = {'a', 'b', 'c'}
b = {'c', 'd', 'e'}

# Union
```

Es la creacion de un nuevo set, uniendo el set principal y el set
a evaluar

```python
c = a.union(b)
print(c)

# Interseccion
```

Son los elementos que coinciden en el set a evaluar

```python
i = a.intersection(b)
print(i)
```

Diferencia

Solo los elementos que no estan presentes en el set a evaluar

```python
diff = a.difference(b)
print(diff)
```
