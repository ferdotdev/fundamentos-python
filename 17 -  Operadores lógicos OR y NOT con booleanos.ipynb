# Operadores de comparaciÃ³n

Estos operadores nos permiten comparar valores y devolver un valor booleano (True o False) los cuales nos indican si la comparacion es verdadera o falsa.

```python
x = 5
y = 3
```

```python
# Igual a
print(x == y)
```

```python
# Diferente de
print(x != y)
```

```python
# Mayor que
print(x > y)
```

```python
# Menor que
print(x < y)
```

```python
# Mayor o igual que
print(x >= y)
```

```python
# Menor o igual que
print(x <= y)
```

## Operadores logicos and, or y not

Estos operadores logicos nos permiten combinar multiples condiciones booleanas y devolver un valor booleano (True o False) basado en la evaluacion de esas condiciones

```python
# AND: Verdadero si ambas condiciones son verdaderas
print(x > 2 and y < 5)
```

```python
# OR: Verdadero si alguna de las condiciones es verdadera
print(x > 2 or y > 5)
```

```python
# NOT: Invierte el valor booleano de la condicion
print(not(x > 2))
```
